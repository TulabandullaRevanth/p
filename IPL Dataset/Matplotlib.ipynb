```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#1 .Plot a line graph to display growth in population in the past 7 decades. Use the following Table Data for this purpose:
census_years = [1951, 1961, 1971, 1981, 1991, 2001, 2011]
population = [361088000, 439235000, 548160000, 683329000, 846387888, 1028737436, 1210726932]
plt.figure(figsize=(10, 6))
plt.plot(census_years, population, marker='o', linestyle='-', color='blue', linewidth=2)
plt.xlabel("Census Year", fontsize=12)
plt.ylabel("Population", fontsize=12)
plt.title("Population Growth Over 7 Decades", fontsize=14, fontweight="bold")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
#2 .Plot a line graph to show Sin Curve.(HOTS) Hint:  Numpy has a function, numpy.sin() to find The cos value
import numpy as np
x = np.linspace(0, 2 * np.pi, 5)  
y = np.sin(x) 
plt.figure(figsize=(10, 5))
plt.plot(x, y, color='blue', linewidth=2, label='sin(x)')
plt.xlabel("x (radians)")
plt.ylabel("sin(x)")
plt.title("Sine Curve", fontsize=14, fontweight="bold")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
#3 . Plot a line graph to show Cos Curve. (HOTS)Hint:Numpy has a function, numpy.cos() to find The cos value
x = np.linspace(0, 2 * np.pi, 5)  
y = np.cos(x)  
plt.figure(figsize=(10, 5))
plt.plot(x, y, color='red', linewidth=2, label='cos(x)')
plt.xlabel("x (radians)")
plt.ylabel("cos(x)")
plt.title("Cosine Curve", fontsize=14, fontweight="bold")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
#4. Plot a Bar Graph to show the number of boys in each class 6- 12. Data should be imagined by student.
classes = ["6", "7", "8", "9", "10", "11", "12"]
boys_count = [45, 50, 48, 52, 47, 40, 38] 
plt.figure(figsize=(10, 6))
plt.bar(classes, boys_count, color='skyblue', alpha=1.0)
plt.xlabel("Class", fontsize=12)
plt.ylabel("Number of Boys", fontsize=12)
plt.title("Number of Boys in Each Class (6-12)", fontsize=14, fontweight="bold")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
#5 .Plot a Bar Graph for Marks scored in different subjects. Data should be imagined.
subjects = ["Math", "Science", "English", "History", "Geography", "Computer"]
marks = [65,88,89,90,40,99] 
plt.figure(figsize=(10, 6))
plt.bar(subjects, marks, color='lightgreen', alpha=0.8)
plt.xlabel("Subjects", fontsize=12)
plt.ylabel("Marks Scored", fontsize=12)
plt.title("Marks Scored in Different Subjects", fontsize=14, fontweight="bold")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
#6. Plot a Histogram to find the number of employees coming to office between 7am to 12 noon.Use bins as 1 hr gaps
arrival_times = [7.10, 7.30, 7.50, 8.0, 8.20, 8.50, 9.0, 9.20, 9.50, 10.0, 10.30, 10.50, 11.0, 11.50, 11.30, 12.0]
plt.figure(figsize=(10, 6))
plt.hist(arrival_times, bins=[7, 8, 9, 10, 11, 12], color='salmon', edgecolor='black', alpha=0.8)
plt.xlabel("Arrival Time (Hour)", fontsize=12)
plt.ylabel("Number of Employees", fontsize=12)
plt.title("Employee Arrival Times (7 AM - 12 Noon)", fontsize=14, fontweight="bold")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()
```
